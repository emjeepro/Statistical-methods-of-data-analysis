---

### Step №1. Create a random variable with normal distribution with parameters m = 5 and σ = 1. Generate 1000 values of this random variable.

### Этап №1. Создать случайную величину, имеющую нормальное распределение с параметрами m = 5 и σ = 1. Сгенерировать 1000 значений этой случайной величины.

```python
# import packages needed for the project

import math
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import scipy.stats as sts
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
normalDist = sts.norm (loc = 5, scale = 1)
normalDist.rvs(size=1000)
```

---

### Step №2. Create a random variable with exponential distribution with parameter λ = 0.2. Generate 1000 values of this random variable.

### Этап №2. Создать случайную величину, имеющую показательное распределение с параметром λ = 0,2. Сгенерировать 1000 значений этой случайной величины.

```python
# for expon in terms of the rate parameter lambda, note that scale is β while λ = 1/β therefore scale = 1/λ

exponentialDist = sts.expon(scale = 5)
exponentialDist.rvs(size=1000)
```

---

### Step №3. Create a random variable with a uniform distribution on [3, 7]. Generate 1000 values of this random variable.

### Этап №3. Создать случайную величину, имеющую равномерное на [3, 7] распределение. Сгенерировать 1000 значений этой случайной величины.

```python
uniformDist = sts.uniform(loc=3, scale=4)
uniformDist.rvs(size=1000)
```

### Этап №4. Для всех последовательностей значений СВ, полученных в пп. 1 – 3, построить нормированные гистограммы (высота столбцов определяется не по частоте, а по относительной частоте попадания в интервал – установить параметр normed = `True`) и графики функций плотности. ### Step №5. Create a random variable with Poisson distribution with parameter a = 5. Generate 1000 values of this random variable. Construct a distribution polygon (by sample) and a graph (point) of the probability function. Place both graphs in one drawing, add a legend.

### Этап №5. Создать случайную величину, имеющую распределение Пуассона с параметром a = 5. Сгенерировать 1000 значений этой случайной величины. Построить многоугольник распределения (по выборке) и график (точечный) функции вероятностей. Оба графика разместить на одном чертеже, добавить легенду.

```python
poissonDist = sts.poisson(5)
poi = poissonDist.rvs(size=1000)

# use method unique() numpy with parameter count True

array = np.array(poi)
unique, counts = np.unique(array, return_counts=True)
dict(zip(unique, counts))
```

```python
# PMF: probability mass function (функция вероятности) : теоретический многоугольник
# лучше добавить эмпирический график вероятностей

x = np.linspace(0,20,21)
pmf = poissonDist.pmf(x)

plt.plot(x, pmf, '-go', label = 'многоугольник')

values = np.unique(poi, return_counts = True)
plt.plot(values[0], values[1]/1000, '-ro', label = 'график вероятностей')
plt.legend(loc="upper right")
plt.show()
```

### Step №6. Generate 30 values of a discrete random variable with the shown below distribution law and construct a statistical series of distribution, compare the values of relative frequencies with theoretical probabilities. Illustrate graphically. Perform the same steps for 300 and 3000 values of the random variable. Draw conclusions.

### Этап №6. Сгенерировать 30 значений дискретной случайной величины, имеющей следующий закон распределения:
| X |  1  |  2  |  3  |  4  |  5  |  6  |
|:-:|:---:|:---:|:---:|:---:|:---:|:---:|
| p | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |

### Построить статистический ряд распределения, сравнить значения относительных частот с теоретическими вероятностями. Проиллюстрировать графически. Выполнить те же действия для 300 и 3000 значений случайной величины. Сделать выводы.

```python
# отрегулируйте ось Y, чтобы лучше визуализировать данные

x = (1, 2, 3, 4, 5, 6)
p = (1/6, 1/6, 1/6, 1/6, 1/6, 1/6)
len_discrete = 30

discrete_rv = sts.rv_discrete(values=(x, p))
discrete_array = discrete_rv.rvs(size=len_discrete)
relative = [list(discrete_array).count(el) / len_discrete for el in x]
print(discrete_array)
plt.plot(x, relative)
plt.figure(1, figsize = (8,3))
plt.ylim(top=0.25)
plt.plot(x, p);
```
