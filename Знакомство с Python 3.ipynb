```python
# суммирование всех введенных пользователем чисел, пока пользователь не введет "end".
S = 0
a = input()
while a!= 'end':
    S += int(a)
    a = input()
print(S)
```

```python
'''
Пользователь вводит пары чисел (компоненты пары - в одной строке через пробел, следующая пара - с новой строки) - 
декартовы координаты точек плскости.
Программа подсчитывает количество точек, попавших в круг заданного радиуса r. 
Подсчет заканчивается как только очередная точка попадает за пределы круга
'''
n = 0
print('Введите радиус круга:')
r = float(input())
print('Вводите координаты точек через пробел:')
while True:
    x, y = input().split()  # разделение по пробелу (по умолчанию)
    x, y = float(x), float(y)
    if (x**2 + y**2 <= r**2):
        n += 1
    else:
        break
print(n)
```

### Определение списка:
#### <имя списка> = [<элемент 1>, <элемент 2>, ... ]
#### [] -
        пустой список.
Доступ к элементам списка - по индексам (начинаются с нуля).   
Индексы могут принимать отрицательные значения. В этом случае номер отсчитывается с конца списка (последний элемент имеет номер -1).   

Элементы списка могут повторяться!

Списочный тип является **изменяемым** - можно явно изменять элементы списка.   
Выводить на печать можно как список целиком, так и отдельные элементы списка.

```python
list = [1, 2, [1, 2, 3], 'Python', True]
print(list)
print(list[1])
print(list[-2])
print(list[2][2])
```

```python
list[2][2] = 5
list[-2] = 'Python 3'
print(list)
```

Можно использовать диапазон значений индексов: в квадратных скобках те же параметры, что у range

```python
print(list[:3]) # вывод всех элементов с 0-го по 2-й
print(list[1:-1]) # вывод всех элементов  1-го по предпоследний (-1 не включается)
print(list[0::2]) # вывод всех элементов списка через 1
print(list[2][1:]) # вывод всех элементов вложенного списка, начиная с 1-го
```

### Особенность работы со списками в Python:
если было выполнено присваивание list2 = list1, то имена обоих списков будут связаны с одной и той же областью памяти,   
поэтому при изменении элементов списка list2 изменится и список list1 и наоборот (!!!);   
но если выполнить переприсвоение всему списку list1, то будет создан новый объект, и его дальнейшие изменения не отразятся на list2

```python
list1 = [1, 2, 3]
list2 = [4, 5]
print('list1 ', list1)
print('list2 ', list2)
list2 = list1
list1[0] = 0
print('list2 после изменения list1', list2)
list2[2] = 5
list2.append(6)
print('list1 после изменения list2', list1)
list1 = [-1, -2]
list1.append(-3)
print('новый list1', list1)
print('list2 после создания нового list1', list2)
```

## Функции
#### Объявление функции
###### def <имя функции>(<список аргументов>):
    <Тело функции>   
    return <список имен переменных или выражений>
#### Вызов функции
    <имя переменной> = <имя функции>(<список параметров>)
    
Функция может не возвращать никакое значение (return может отсутствовать либо после return - пустая строка); в этом случае выполняется набор действий, описанный в теле функции. Нужно вызывать как метод (без = ).

Функция может быть без параметров: <имя функции>()    
Функция может иметь произвольное число параметров (пример - print). Определение такой функции:    
**def**<имя функции>(*<имя последовательности, в которой будут храниться параметры>):   
К элементам последовательности параметров можно обращаться по индексу.
#### Определение значений параметров по умолчанию
###### def <имя функции>(<аргумент 1>, <аргумент 2>, ... , <аргумент k> = <значение по умолчанию>,  ... ):
#####    
Конструкцию вида <аргумент> = <значение> можно использовать и при вызове функции. В этом случае можно нарушать порядок слеедования аргументов.
##### Локальные переменные
Если имя локальной переменной совпадает с именем глобальной переменной, то изменение значения локальной переменной **не повлияет** на значение глобальной переменной. Отсюда следует, что **нельзя** с помощью функции изменить значение глобальной переменной! Можно изменить список, например, добавить новый элемент (т. к. списочный тип - изменяемый). Но! Если в теле функции создать новый список, то связь со старым разрушится.

## Словари
Позволяют хранить множество пар вида (ключ, значение).   
Элементы словаря (т. е. пары (ключ, значение)) **не упорядочены**.  

Все ключи должны быть различны: при добавлении в словарь нового элемента с тем же ключом старое значение затирается.    
Ключи должны быть неизменяемого типа (могут быть числами, строками, но **не** списками, **не** др. словарями).   
Одному ключу может быть сопоставлен список значений.
Например:    
    (№ автомобиля, ФИО владельца). По ключу "№ автомобиля" (имеет уникальное значение) можно при необходимости определить ФИО владельца.  
А может быть (№ автомобиля, список владельцев).

Значения ключей в одном словаре могут быть разных типов.  
   
Для словарей используется тип данных **dict**.    
Этот тип данных - **изменяемый** (можно изменять словарь в процессе обработки).   
Доступ к элементам словаря - по ключу.
### Определение словаря
##### Определение пустого словаря - <имя словаря> = {}
##### Определение непустого словаря - <имя словаря> = {<ключ 1>:<список значений 1>, <кюч 2>:<список значений 2>,  ...  }
### Операции со словарями
#### Проверка, имеется ли данный ключ в словаре:
    <ключ> in <имя словаря> - возвращает True или False
    <ключ> not in <имя словаря> 
#### Добавление нового элемента:
    <имя словаря>[<ключ>] = <значение/список значений>
#### Добавление в список значений уже существующего ключа нового значения:
    <имя словаря>[<ключ>] += [<значение/список значений>]  или   
    <имя словаря>[<ключ>].append[<значение/список значений>]
#### Удаление элемента из словаря:
    del <имя словаря>[<ключ>]
#### Поиск значения по ключу:
    <имя словаря>.get[<ключ>] - возвращает значение ключа или None, если такого ключа нет
### Поэлементная обработка словаря в цикле for
#### по ключам:
##### for <ключ> in <имя словаря>:
    <Блок операторов>
##### for <ключ> in <имя словаря>.keys():
    <Блок операторов>
#### по значениям:
##### for <значение>  in <имя словаря>.values():
    <Блок операторов>
#### по парам:
##### for <ключ>, <значение>  in <имя словаря>.items():
    <Блок операторов>

```python
# Словарь содержит результаты сессии по группе (на текущий момент) - оценки без привязки к дисциплинам (для упрощения)
d = {'Иванов': [3, 4, 5], 'Петров':[5, 5], 'Сидоров': [3]}
print(d)
```

```python
# Добавление новой оценки конкретному студенту
d['Сидоров'] += [3]
print(d)
```

```python
# Добавление информации о новом студенте (нового элемента словаря)
d['Васильев'] = [2, 2, 2]
print(d)
```

```python
# Удаление информации о неуспевающем студенте (после отчисления)
del d['Васильев']
print(d)
```

```python
# Вывод информации по конкретному студенту
print(d['Петров'])
print(d.get('Петров'))
print(d.get('Васильев'))
# следующая команда вызовет ошибку
# print(d['Васильев'])
```

```python
for k, v in d.items():
    print(k, v)
```
