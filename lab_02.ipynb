```python
# import packages needed for the project

import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as sts
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# for expon in terms of the rate parameter lambda, note that scale is β while λ = 1/β therefore scale = 1/λ

λ = 0.2
exp = sts.expon(scale = 1/λ)

# generate random numbers
exponentialDist = exp.rvs(1000)

fig = plt.figure()
draw = fig.add_axes([0., 0., 1., 1.])

draw.hist(exponentialDist, density = True, bins = 22, label = 'показательное распределение', ec="white")

# matplotlib.axes.Axes.set_xlim
draw.set_xlim(0, 20)

x = np.linspace(0, 20, 100)
pdf = exp.pdf(x)
draw.plot(x, pdf, 'r', label = 'график плотности')
draw.legend(loc='upper right');
```

central limit theorem

```python
# rate parameter for the exponentially distributed population
rate = 0.2

# population mean
mu = 1/rate

# population standard deviation
sd = np.sqrt(1/(rate**2))

print('mean:', mu)
print('standard deviation:', sd)
```

```python
# drawing 1000 random samples of size 5 from the exponentially distributed population

sample_size = 5
df5 = pd.DataFrame(index= ['x1', 'x2', 'x3', 'x4', 'x5'] )

for i in range(1, 1001):
    exponential_sample = np.random.exponential((1/rate), sample_size)
    col = f's{i}'
    df5[col] = exponential_sample

# taking a peek at the samples
df5
```

```python
# calculating sample means and plotting their distribution

df5_sample_means = df5.mean()
plt.hist(df5_sample_means, density=True, stacked=True, bins=15, color="purple", label="'mean' каждой выборки", ec="white")
plt.legend(loc='upper right');
```

```python
# the first 5 values from the 1000 sample means

df5_sample_means.head()
```

```python
# an estimate of the mean of the sampling distribution can be obtained as:

np.mean(df5_sample_means)
```

#### The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacement , then the distribution of the sample means will be approximately normally distributed.

#### Центральная предельная теорема утверждает, что если у вас есть генеральная совокупность со средним значением μ и стандартным отклонением σ и вы берете достаточно большие случайные выборки из генеральной совокупности с заменой, то распределение выборочных средних будет приблизительно нормально распределенным.

$σ_\overline{x} = \frac{\mathrm{σ} }{\mathrm{\sqrt{n}}}$

```python
# the above value is very close to the value stated by the CLT, which is:

CLT = sd / np.sqrt(sample_size)
print(CLT)
```

```python
plt.hist(mean, density=True, stacked=True, bins=15, color="orange", label="'mean' каждой выборки", ec="white")
plt.axvline(CLT, color='k', linestyle='dashed', linewidth=1, label="центральная предельная теорема")
min_ylim, max_ylim = plt.ylim()
plt.text(CLT*1.1, max_ylim*0.9, 'CLT: {:.2f}'.format(CLT))
plt.legend(loc='upper right');
```
